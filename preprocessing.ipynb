```python
generations = [
    (df['#'] <= 151),                    # GENERATION 1
    (df['#'] > 151) & (df['#'] <= 251),  # GENERATION 2
    (df['#'] > 251) & (df['#'] <= 386),  # GENERATION 3
    (df['#'] > 386) & (df['#'] <= 493),  # GENERATION 4
    (df['#'] > 493) & (df['#'] <= 649),  # GENERATION 5
    (df['#'] > 649) & (df['#'] <= 721),  # GENERATION 6
    (df['#'] > 721) & (df['#'] <= 809),  # GENERATION 7
    (df['#'] > 809) & (df['#'] <= 905),  # GENERATION 8
    (df['#'] > 905) & (df['#'] <= 1008), # GENERATION 9
    ]
```

```python
values = ["Generation 1", "Generation 2", "Generation 3", "Generation 4", "Generation 5", "Generation 6", "Generation 7", "Generation 8", "Generation 9"]
```

```python
df['Generation'] = np.select(generations, values)
```
